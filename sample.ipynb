```python
# Step 2: Set up environment variables for API keys
import os



# Set your Gemini API key (from https://makersuite.google.com/app/apikey)
os.environ["GOOGLE_API_KEY"] = "your_google_gemini_api_key"

# OpenRouter: Get your free key from https://openrouter.ai/
# OpenRouter is a universal API gateway that lets you access many LLMs (like DeepSeek) using a single OpenAI-style interface.
# It’s great for switching between models easily, often with free access to open-source models.

os.environ["OPENROUTER_API_KEY"] = "your_openrouter_api_key"
```

```python
# Step 3: Import litellm and enable debug logs (optional)
import litellm
```

```python
# Step 4: Define a unified prompt function
def ask_model(model_name, prompt, **kwargs):
    response = litellm.completion(
        model=model_name,
        messages=[{"role": "user", "content": prompt}],
        **kwargs
    )
    return response['choices'][0]['message']['content']
```

```python
#  Example 1: Using Gemini
response_gemini = ask_model(
    model_name="gemini/gemini-1.5-flash",
    prompt="What are the benefits of eating vegetables?",
    api_key= os.environ["GOOGLE_API_KEY"]
)
print("Gemini Response:\n", response_gemini)
```

```python
#  Example 2: Using Deepseek
response_deepseek = ask_model(
    model_name="openrouter/deepseek/deepseek-chat",
    prompt="What are the benefits of eating vegetables?",
    api_key= os.environ["OPENROUTER_API_KEY"]
)
print("DeepSeek (OpenRouter) Response:\n", response_deepseek)
```

## Evaluation

```python
# Step 1: Import
from sentence_transformers import SentenceTransformer, util

# Step 2: Load a Vietnamese embedding model
model = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")

# Step 3: Define a scoring function
def is_match(label: str, prediction: str, threshold: float = 0.8):
    embeddings = model.encode([label, prediction], convert_to_tensor=True)
    similarity = util.cos_sim(embeddings[0], embeddings[1]).item()
    print(f"Similarity: {similarity:.4f}")
    return similarity >= threshold

# Step 4: Try with a test case
label = "Có"
prediction = "Dạ có"

match = is_match(label, prediction)
print("✅ Match!" if match else "❌ Not a match.")
```

```python
response_gemini = ask_model(
    model_name="gemini/gemini-1.5-flash",
    prompt="Thủ đô của Việt Nam là?",
    api_key= os.environ["GOOGLE_API_KEY"]
)

label = "Thủ đô của Việt Nam là thành phố Hà Nội"


match = is_match(label, response_gemini)
print(response_gemini)
print("✅ Match!" if match else "❌ Not a match.")
```
