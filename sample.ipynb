```python
buffered = BytesIO()
image.save(buffered, format="JPEG")
img_base64 = base64.b64encode(buffered.getvalue()).decode()

os.environ["GEMINI_API_KEY"] = "AIzaSyCoUaLN9E05l6-sFrRRMTbxslfC9Hd_9yc"

prompt = "What's in the picture?"

messages = [
    {
        "role": "user",
        "content": [
            {"type": "text", "text": prompt},
            {
                "type": "image_url",
                "image_url": {
                    "url": f"data:image/jpeg;base64,{img_base64}"
                }
            }
        ]
    }
]

response = litellm.completion(
    model="gemini/gemini-2.5-pro-exp-03-25",
    messages=messages
)

content = response.get("choices", [{}])[0].get("message", {}).get("content")
print(content)
```

## Evaluation

```python
# Step 1: Import
from sentence_transformers import SentenceTransformer, util

# Step 2: Load a Vietnamese embedding model
model = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")

# Step 3: Define a scoring function
def is_match(label: str, prediction: str, threshold: float = 0.8):
    embeddings = model.encode([label, prediction], convert_to_tensor=True)
    similarity = util.cos_sim(embeddings[0], embeddings[1]).item()
    print(f"Similarity: {similarity:.4f}")
    return similarity >= threshold

# Step 4: Try with a test case
label = "Có"
prediction = "Dạ có"

match = is_match(label, prediction)
print("✅ Match!" if match else "❌ Not a match.")
```
